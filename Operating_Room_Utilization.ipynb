Time difference between OR Schedule and Wheels In for each schedule

```python
df['OR Schedule']=pd.to_datetime(df['OR Schedule'])
df['Wheels In']=pd.to_datetime(df['Wheels In'])
df['time_bet_orschedule_wheelsin(min)']=abs(df['Wheels In']-df['OR Schedule']).dt.total_seconds()/60
df.head()
```

Time difference between Start Time and Wheels In Time for each schedule

```python
df['Start Time']=pd.to_datetime(df['Start Time'])
df['Wheels In']=pd.to_datetime(df['Wheels In'])
df['time_bet_wheelsin_starttime(min)']=abs(df['Start Time']-df['Wheels In']).dt.total_seconds()/60
df.head()
```

Time difference between Start Time and End Time for each schedule

```python
df['Start Time']=pd.to_datetime(df['Start Time'])
df['End Time']=pd.to_datetime(df['End Time'])
df['time_bet_start_endtime(min)']=abs(df['End Time']-df['Start Time']).dt.total_seconds()/60
df.head()
```

Time difference between End Time and Wheels Out for each schedule

```python
df['End Time']=pd.to_datetime(df['End Time'])
df['Wheels Out']=pd.to_datetime(df['Wheels Out'])
df['time_bet_endtime_wheelsout(min)']=abs(df['Wheels Out']-df['End Time']).dt.total_seconds()/60
df.head()
```

Time difference between Wheels In and Wheels Out for each schedule

```python
df['Wheels In']=pd.to_datetime(df['Wheels In'])
df['Wheels Out']=pd.to_datetime(df['Wheels Out'])
df['time_bet_wheelsin_wheelsout(min)']=abs(df['Wheels Out']-df['Wheels In']).dt.total_seconds()/60
df.head()
```

Overall Utilization Rates and Utilization Rate for Each Week: Overall Utilization: Calculate the total time the OR was used versus the total available OR time. Weekly Utilization: Determine utilization rates for each week to identify trends or fluctuations over time.

```python
total_available_time=df['Booked Time (min)'].sum()
total_utilized_time=df['time_bet_start_endtime(min)'].sum()
utilization_rate=total_utilized_time/total_available_time
utilization_rate
```

```python
column=['Date','Booked Time (min)','time_bet_start_endtime(min)']
df_weekly=df[column]
df_weekly['Date'] = pd.to_datetime(df_weekly['Date'])
df_weekly.set_index('Date', inplace=True)
df_weekly = df_weekly.resample('W').sum()
df_weekly['Utilization Rate'] = df_weekly['time_bet_start_endtime(min)'] / df_weekly['Booked Time (min)']
df_weekly
```

Time difference between the booked time and actual utilized time and number of time OR has been overbooked and underbooked

```python
df['diff_booking_usetime']=abs(df['Booked Time (min)']-df['time_bet_start_endtime(min)'])
df['Comparision_booking_usetime(start_and_end)']=np.select([df['time_bet_start_endtime(min)']<df['Booked Time (min)'],df['time_bet_start_endtime(min)']>df['Booked Time (min)'],
                                                      df['time_bet_start_endtime(min)']==df['Booked Time (min)']],['More','Less','Equal'])
df['Comparision_booking_usetime(start_and_end)'].value_counts()
```
