# **Pre-processing of Data**

```python
import seaborn as sns
sns.heatmap(data=df.isnull(), cmap='viridis')
```

```python
#Unnesting columns of dataset
#Director column
df_split = df['director'].str.split(',')
df_new1=pd.DataFrame(df_split)
df_new1=df_new1.set_index(df['title'])
df_new1=df_new1.explode('director')
df_new1.head()
```

```python
#Unnesting columns of dataset
#Cast column
df_split = df['cast'].str.split(',')
df_new2=pd.DataFrame(df_split)
df_new2=df_new2.set_index(df['title'])
df_new2=df_new2.explode('cast')
df_new2.head()
```

```python
#Unnesting columns of dataset
#Country column
df_split = df['country'].str.split(',')
df_new3=pd.DataFrame(df_split)
df_new3=df_new3.set_index(df['title'])
df_new3=df_new3.explode('country')
df_new3.head()
```

```python
#Unnesting columns of dataset
#Listed_in column
df_split = df['listed_in'].str.split(',')
df_new4=pd.DataFrame(df_split)
df_new4=df_new4.set_index(df['title'])
df_new4=df_new4.explode('listed_in')
df_new4.head()
```

```python
merged_df=pd.merge(df_new1,df_new2,on='title',how='outer')
merged_df1=pd.merge(df_new3,df_new4,on='title',how='outer')
df_new=pd.merge(merged_df,merged_df1,on='title',how='outer')
df_new
```

```python
#Merging the unnested columns with the original column
df_final=pd.merge(df_new,df[['title','show_id','type','date_added','release_year','rating','duration',	'description']],on='title',how='left')
df_final
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot using Seaborn
plt.figure(figsize=(15, 6))  # Adjust the figure size if needed
sns.countplot(data=df_final, x='listed_in')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

plt.show()
```
