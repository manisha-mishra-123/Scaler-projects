```python
from google.colab import files
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import string
```

```python
df=pd.read_excel("Input.xlsx")
df
```

```python
StopWords_Auditor=pd.read_csv("StopWords_Auditor.txt",header=None)
StopWords_Auditor[0]=StopWords_Auditor[0].str.lower()

StopWords_Currencies= pd.read_csv("StopWords_Currencies.csv",header=None)
StopWords_Currencies[0]=StopWords_Currencies[0].str.lower()

StopWords_DatesandNumbers=pd.read_csv('StopWords_DatesandNumbers.txt',header=None)
StopWords_DatesandNumbers[0]=StopWords_DatesandNumbers[0].str.lower()

StopWords_Generic=pd.read_csv('StopWords_Generic.txt',header=None)
StopWords_Generic[0]=StopWords_Generic[0].str.lower()

StopWords_GenericLong=pd.read_csv('StopWords_GenericLong.txt',header=None)
StopWords_GenericLong[0]=StopWords_GenericLong[0].str.lower()

StopWords_Geographic=pd.read_csv('StopWords_Geographic.txt',header=None)
StopWords_Geographic[0]=StopWords_Geographic[0].str.lower()

StopWords_Names=pd.read_csv('StopWords_Names.txt',header=None)
StopWords_Names[0]=StopWords_Names[0].str.lower()

positive_word=pd.read_csv('positive-words.txt',header=None)

negative_word=pd.read_csv('negative words.csv',header=None)
```

```python
for n in range(len(df)):
  page=requests.get(df['URL'][n])
  if page.status_code==200:
    soup=BeautifulSoup(page.text,'html.parser')
    title=soup.title.text
    target_div=soup.find('div', class_="td-post-content tagdiv-type")
    if target_div is not None:
      target=target_div.get_text(separator=' ',strip=True)
      text = f"{title}\n\n{target}"
    #print(text,"\n")
  filtered_text=""
  word=0
  complex_word=0
  s=text.split()
  if text is not None:
    for i in s:
      translator = str.maketrans("", "", string.punctuation)
      cleaned_word = i.translate(translator)
      if ~(StopWords_Auditor[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_Currencies[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_DatesandNumbers[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_Generic[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_GenericLong[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_Geographic[0].str.contains(cleaned_word.lower())).any() and ~(StopWords_Names[0].str.contains(cleaned_word.lower())).any():
        filtered_text+=i+" "
        word+=1
        syllables=0
        for j in cleaned_word:
          if j.lower()=='a' or j.lower()=='e' or j.lower()=='i' or j.lower()=='o' or j.lower()=='u':
            syllables+=1
        if i.endswith(('es','ed')):
          syllables-=1
        if syllables>2:
          complex_word+=1
  #print(filtered_text,"\n")
  #print(word, complex_word)
  positive_score=0
  negative_score=0
  t=filtered_text.split()
  for a in t:
    translator = str.maketrans("", "", string.punctuation)
    cleaned_word = a.translate(translator)
    if positive_word[0].str.contains(cleaned_word).any():
      positive_score+=1
    if negative_word[0].str.contains(cleaned_word).any():
      negative_score+=1

  #print(positive_score)
  #print(negative_score)

  df.at[n, "Positive Score"]=positive_score
  df.at[n, "Negative Score"]=negative_score

  polarity_score=(positive_score-negative_score)/((positive_score+negative_score)+0.000001)
  #print(polarity_score)

  df.at[n, "Polarity Score"]=polarity_score

  subjectivity_score = (positive_score + negative_score)/ (word + 0.000001)
  #print(subjectivity_score)

  df.at[n, "Subjectivity Score"]=subjectivity_score

  char=0
  sen=0
  for i in filtered_text:
    if i==".":
      sen+=1
    char+=1

  average_sentence_length=word/sen
  Percentage_of_Complex_words=complex_word/word
  Fog_Index = 0.4 * (average_sentence_length + Percentage_of_Complex_words)

  df.at[n, "AVG SENTENCE LENGTH"]=average_sentence_length
  df.at[n, "PERCENTAGE OF COMPLEX WORDS"]=Percentage_of_Complex_words
  df.at[n, "FOG INDEX"]=Fog_Index

  df.at[n, "Average Number of Words Per Sentence"]=average_sentence_length

  df.at[n, "COMPLEX WORD COUNT"]=complex_word
  df.at[n, "WORD COUNT"]=word

  #df.at[n, "SYLLABLE PER WORD"]=total_syllables

  pronounRegex = re.compile(r'\b(?:I|we|my|ours|us)\b', re.I)
  pronouns = len(pronounRegex.findall(filtered_text))

  df.at[n, "PERSONAL PRONOUNS"]=pronouns

  Average_Word_Length= char/word

  df.at[n, "Average Word Length"]=Average_Word_Length

df
```

Positive Score: This score is calculated by assigning the value of +1 for each word if found in the Positive Dictionary and then adding up all the values.

Negative Score: This score is calculated by assigning the value of -1 for each word if found in the Negative Dictionary and then adding up all the values. We multiply the score with -1 so that the score is a positive number.

Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)


Subjectivity Score: This is the score that determines if a given text is objective or subjective. It is calculated by using the formula:

Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
